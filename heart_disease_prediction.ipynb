```python
import pandas as pd
import numpy as np
import csv
from sklearn.utils import shuffle
```

```python
df = pd.read_csv('complete.csv')
df= df.replace('?',np.nan)
df = df.astype(float)
df.fillna(df.median(),inplace=True)
df = df.astype(int)
df = shuffle(df)
cols = df.columns
col_train = cols[:len(cols)-1]
x = df[col_train]
y= df['predict']
x.tail()
```

```python
from sklearn import cross_validation,neighbors,svm,tree 

from sklearn.utils import shuffle

from sklearn.ensemble import RandomForestClassifier

X_train,X_test,y_train,y_test = cross_validation.train_test_split(x,y,test_size=0.2) 

clf = svm.LinearSVC() 
clf1 = svm.SVC() 
clf2 = neighbors.KNeighborsClassifier()
clf3 = RandomForestClassifier()


clf.fit(X_train,y_train) 
clf1.fit(X_train,y_train) 
clf2.fit(X_train,y_train) 
clf3.fit(X_train,y_train)


score = clf.score(X_test,y_test) 
score1 = clf1.score(X_test,y_test) 
score2  = clf2.score(X_test,y_test)
score3  = clf3.score(X_test,y_test)


print "Accuracy of LinearSVC is " + str(score) 
print "Accuracy of SVC is " + str(score1)
print "Accuracy of KNN is " + str(score2)
print "Accuracy of RandomForest is " + str(score3)

# Random forest has highest accuracy to predict the correct result
```

```python
# Percentage of belonging to each label
# Use Random Forest
val1=0 
val2=0 
val3=0 
val4=0 
val5=0


for predict in clf3.predict(X_test):    
    if predict==0:
        val1+=1
    elif predict==1:
        val2+=1
    elif predict==2:
        val3+=1
    elif predict==3:
        val4+=1
    elif predict==4:
        val5+=1  
        
        
print "Percentage of belonging to label 0: "+str((val1*1.0)/len(X_test))
print "Percentage of belonging to label 1: "+str((val2*1.0)/len(X_test))
print "Percentage of belonging to label 2: "+str((val3*1.0)/len(X_test))
print "Percentage of belonging to label 3: "+str((val4*1.0)/len(X_test))
print "Percentage of belonging to label 4: "+str((val5*1.0)/len(X_test))


```
